```python
amountInZip= dict()
for index, row in dataset.iterrows():
    zip = row['OwnerZip']
    
    if zip in amountInZip:
        amountInZip[zip] = amountInZip[zip] + 1
    else:
        amountInZip[zip] = 1
print(amountInZip)
```
