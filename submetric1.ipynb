```python
import pandas as pd
data1 = pd.read_csv("data1.csv", sep=",")
data1.sample(10)
```

```python
amountInZip=dict()
for index, row in data1.iterrows():
    zip = row['incident_zip']
    if zip in amountInZip:
        amountInZip[zip]=amountInZip[zip]+1
    else:
        amountInZip[zip]=1
print(amountInZip)
```
