# **Introduction**

For our final project we wanted to figure out what the best neighborhood was in Pittsburgh. We went onto WPRDC to gather 3 data sets to compare in order to answer our question. We believe that the friendliest neighborhood is considered the best neighborhood so that was the metric we used.

# **Metric**

Our overall metric for the project was friendliest. We broke that down into 3 submetrics. The first was amount of overdoeses in neighborhoods in Pittsburgh. We looked for neighborhoods with low overdoses because we thought that people overdosing was not very friendly. The second submetric was the amount of dog licenses in neighborhoods in Pittsburgh. We looked for neighborhoods with high dog lisences because we love dogs and think it makes neighborhoods more friendly. The last submetric we used was amount of private schools in neighborhoods in Pittsburgh. We looked for a low amount of private schools in a neighborhood because we think private school kids are stuck up which makes the neighborhood less friendly. The data set we used for drug overdoses was https://data.wprdc.org/dataset/allegheny-county-fatal-accidental-overdoses . The data set we used for dog lisences was https://data.wprdc.org/dataset/allegheny-county-dog-licenses . The data set we used for private schools was https://data.wprdc.org/dataset/allegheny-county-private-schools-locations . 

# **Data**

## **Overdoses**

```python
import pandas as pd
data1 = pd.read_csv("data1.csv", sep=",")
amountInZip=dict()
for index, row in data1.iterrows():
    zip = row['incident_zip']
    if zip in amountInZip:
        amountInZip[zip]=amountInZip[zip]+1
    else:
        amountInZip[zip]=1
print(amountInZip)
```

## **Dog Licenses**

```python
import pandas as pd
dataset = pd.read_csv("LifetimeDogLicenses.csv", sep=",")
amountInZip= dict()
for index, row in dataset.iterrows():
    zip = row['OwnerZip']
    
    if zip in amountInZip:
        amountInZip[zip] = amountInZip[zip] + 1
    else:
        amountInZip[zip] = 1
print(amountInZip)
```

## **Private Schools**

```python
import pandas as pd
data3 = pd.read_csv("data3.csv", sep=",")
numberOfSchools=dict()
for index, row in data3.iterrows():
    zip = row['location_zip_code']
    if zip in numberOfSchools:
        numberOfSchools[zip]=numberOfSchools[zip]+1
    else:
        numberOfSchools[zip]=1
print(numberOfSchools)
```
